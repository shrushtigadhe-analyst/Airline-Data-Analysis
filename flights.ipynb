```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv(r"C:\Users\Shrushti Gadhu\Downloads\archive (1)\flights.csv")
```

```python
data['flight'].value_counts().plot.bar()
plt.title("flight per name")
plt.ylabel("Number of flight")
plt.show()
```

```python
%matplotlib inline
data['time_hour'] = pd.to_datetime(data['time_hour'])
data.groupby('time_hour').size().plot.line()
plt.title("flights Over hour ")
plt.ylabel("Number of flight")
plt.show()

```

```python
import matplotlib.pyplot as plt
%matplotlib inline
data['arr_delay'].plot.hist(bins=30)
plt.title("Distribution of arr_delay")
plt.xlabel("arr_delay (minutes)")
plt.show()
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
data['time_hour'] = pd.to_datetime(data['time_hour'])
hourly = data.groupby(data['time_hour'].dt.hour).size()
hourly.plot(kind='line', color='skyblue', marker='o')
plt.title("flight per Hour")
plt.xlabel("hour of day")
plt.ylabel("Number of flight")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(8,5))
sns.histplot(data['arr_delay'], bins=50, kde=True, color='steelblue')
plt.title("Distribution of arr_delay")
plt.xlabel("arr_delay (minute)")
plt.xlim(-50, 300)  # Focus on realistic range
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
revenue = data.groupby('name')['year'].sum().sort_values(ascending=False)
revenue.plot(kind='bar', color='mediumseagreen')
plt.title("year by name")
plt.ylabel("year")
plt.xticks(rotation=45)
plt.show()

```

```python
import matplotlib.pyplot as plt
%matplotlib inline
top_passengers = data.groupby('flight')['id'].sum().nlargest(10)
top_passengers.plot(kind='bar', color='orange')
plt.title("Top 10 name by id")
plt.ylabel("id")
plt.xticks(rotation=45)
plt.show()
```
